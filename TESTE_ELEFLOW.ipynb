### O qual a diferença entre um aprendizado de máquina supervisionado e um não supervisionado?

Dê exemplos de algorítimos para cada um.

---

A aprendizagem de máquina é classificada principalmente em dois tipos: supervisionada e não supervisionada.

#### **Aprendizado Supervisionado:**

No aprendizado supervisionado, os algoritmos de aprendizado de máquina são treinados usando conjuntos de dados rotulados. Isso significa que a resposta ou o resultado desejado já é conhecido. Durante o treinamento, o algoritmo procura padrões nos recursos que estão associados aos rótulos. Uma vez treinados, estes modelos são então aplicados a novos dados para prever resultados.

Exemplos de algoritmos de aprendizado supervisionado incluem:

- Regressão linear e logística
- Máquinas de vetores de suporte (SVM)
- Árvores de decisão
- Florestas aleatórias
- Redes neurais profundas (quando possuem dados rotulados)

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn import metrics
```

```python
# choose "Year" and "NA_Sales" columns as features
features = df[['Year', 'NA_Sales']]

# choose "Global_Sales" column as target
target = df['Global_Sales']
```

O modelo de Regressão Linear do Scikit-learn realmente não pode lidar com valores `NaN`. Você precisa limpar ou preencher esses valores faltantes antes de usar o modelo.

Vamos usar a técnica de imputação para preencher os valores faltantes. A imputação é uma técnica convencional para lidar com dados faltantes. Neste caso, vou usar `SimpleImputer` do módulo `sklearn.impute` para preencher os valores faltantes com a mediana:

```python
# create an imputer object that fills 'NaN' values with the  median
imputer = SimpleImputer(missing_values=np.nan, strategy='median')

df[['Year', 'NA_Sales']] = imputer.fit_transform(df[['Year', 'NA_Sales']])

# choose "Year" and "NA_Sales" columns as features
features = df[['Year', 'NA_Sales']]

# choose "Global_Sales" column as target
target = df['Global_Sales']

# Split dataset into training set and test set
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.2, random_state=1)

# Create a Linear Regression model
regressor = LinearRegression()

# Train the model using the training sets
regressor.fit(features_train, target_train)

# Make predictions using the testing set
target_pred = regressor.predict(features_test)

# Print model performance
print('Mean Absolute Error:', metrics.mean_absolute_error(target_test, target_pred))
print('Mean Squared Error:', metrics.mean_squared_error(target_test, target_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(target_test, target_pred)))
```

**Nota:** Este é um exemplo bem simplificado. Na prática, você provavelmente precisaria fazer um pouco de limpeza/transformação de dados, seleção de características e ajuste de modelo para obter um modelo de regressão linear que realmente funcione bem em seus dados. Além disso, a regressão linear pode não ser o melhor modelo para este tipo de previsão, dependendo da natureza dos seus dados. Além disso, a escolha do recurso ('Year' e 'NA_Sales') no exemplo é aleatória e pode não ser a melhor escolha.


---
#### **Aprendizado Não Supervisionado:**

No aprendizado não supervisionado, os algoritmos de aprendizado de máquina são treinados usando conjuntos de dados não rotulados. O algoritmo tenta identificar padrões e relações nos dados, mesmo que não saibamos quais são as respostas corretas.

Em outras palavras, ele aprende os padrões nos dados e, com base nesses padrões, ele pode agrupar ou associar novos dados.

Exemplos de algoritmos de aprendizado não supervisionado incluem:

- Clustering (por exemplo, algoritmo K-means, Clustering Aglomerativo Hierárquico)
- Análise de componentes principais (PCA)
- Regras de associação (por exemplo, Apriori, FP-growth)
- Autoencoders em redes neurais (tipo especial de rede neural usada para representação de dados e detecção de anomalias).

A técnica de Clustering, também conhecida como agrupamento, pode ser uma ferramenta eficaz na detecção de padrões e tendências em conjuntos de dados não rotulados. Quando aplicada ao conjunto de dados de video games citado anteriormente, diversas abordagens estratégicas podem ser consideradas:

* **Agrupamento baseado em vendas**: é possível agrupar jogos de acordo com seu desempenho de vendas. Os recursos para isso podem ser as vendas globais, vendas na América do Norte, Europa, entre outros. Isso potencialmente revela agrupamentos de jogos com desempenho de vendas semelhante.

* **Agrupamento baseado em gêneros de jogos**: a análise pode ser direcionada às características dos jogos, como gênero, plataforma etc. A partir disso, é possível formar agrupamentos de jogos que compartilham semelhanças nestes aspectos.

* **Agrupamento baseado em avaliações**: ao ter disponíveis avaliações de usuários ou críticas, o agrupamento pode ajudar a identificar categorias de jogos que geram respostas semelhantes do público ou da crítica.

Para executar o agrupamento, certos métodos podem ser utilizados. Dentre eles estão o K-means, DBSCAN ou o agrupamento hierárquico aglomerativo. A escolha do método ideal pode depender da natureza dos dados e do problema em questão.

Ao optar pelo algoritmo K-means, um dos mais populares para agrupamento, é necessário predefinir a quantidade de agrupamentos (k) a serem formados. Em seguida, o algoritmo designa cada ponto ao agrupamento cujo centro, ou "centroid", esteja mais próximo. O "centroid" de um agrupamento é uma combinação de recursos que define o "centro" do agrupamento.

Importante ressaltar que os métodos de agrupamento necessitam de algum pré-processamento dos dados, incluindo a normalização ou padronização dos recursos e o tratamento dos valores ausentes. No caso das medidas de similaridade baseadas em distância, como o K-means, uma escala unificada para todos os recursos é crucial para a acurácia dos agrupamentos formados.

**Nota:** Existe um terceiro tipo de aprendizado de máquina conhecido como aprendizado por reforço, no qual um agente aprende como se comportar em um ambiente executando certas ações e observando os resultados.

Analisando apenas essas matrizes qual modelo você considera o melhor para ser utilizado? Justifique sua resposta.

Essa pergunta é um pouco difícil de responder sem mais contexto sobre a importância relativa ao classificar corretamente casos positivos e negativos. Todavia, podemos analisar algumas métricas como precisão, recall e acurácia.

No Modelo 1, temos:
- Verdadeiros Positivos (VP) = 9413
- Verdadeiros Negativos (VN) = 5740
- Falsos Positivos (FP) = 519
- Falsos Negativos (FN) = 1119

No Modelo 2, temos:
- Verdadeiros Positivos (VP) = 7330
- Verdadeiros Negativos (VN) = 6751
- Falsos Positivos (FP) = 705
- Falsos Negativos (FN) = 2005

Os cálculos para precisão, recall e acurácia são:

- Precisão = VP / (VP + FP)
- Recall = VP / (VP + FN)
- Acurácia = (VP + VN) / (VP + VN + FP + FN)

Se calcularmos estas métricas para ambos os modelos, temos:

- Modelo 1: Precisão = 0.948, Recall = 0.894, Acurácia = 0.902
- Modelo 2: Precisão = 0.912, Recall = 0.785, Acurácia = 0.839

Com base nessas métricas, o Modelo 1 tem uma maior precisão (propensão a menos falsos positivos), recall (propensão a menos falsos negativos) e acurácia geral, portanto, se estas são as métricas que nos interessam, o Modelo 1 seria o preferido.
Porém, cabe ressaltar que essa conclusão depende da importância que damos a cada tipo de erro em nosso problema específico.

```python
# Função para calcular as estatísticas
def calculate_stats(VP, VN, FP, FN):
    precisao = VP / (VP + FP)
    recall = VP / (VP + FN)
    acuracia = (VP + VN) / (VP + VN + FP + FN)

    return precisao, recall, acuracia

# Modelo 1
VP1, VN1, FP1, FN1 = 9413, 5740, 519, 1119
precisao1, recall1, acuracia1 = calculate_stats(VP1, VN1, FP1, FN1)

# Modelo 2
VP2, VN2, FP2, FN2 = 7330, 6751, 705, 2005
precisao2, recall2, acuracia2 = calculate_stats(VP2, VN2, FP2, FN2)

# Imprimir os resultados
print("Modelo 1: Precisão = {:.3f}, Recall = {:.3f}, Acurácia = {:.3f}".format(precisao1, recall1, acuracia1))
print("Modelo 2: Precisão = {:.3f}, Recall = {:.3f}, Acurácia = {:.3f}".format(precisao2, recall2, acuracia2))
```

Além disso, é importante destacar que, nessa situação, talvez o modelo de classificação não seja o mais indicado.


Ainda neste caso, estamos lidando com um problema de classificação multiclasse. As métricas de Precisão e Recall são originalmente definidas para problemas de classificação binária.

A curva Precision-Recall (PR curve) é um gráfico que plota a precisão contra o recall para cada limite de probabilidade. Isso é comumente usado em problemas de classificação binária para entender o trade-off entre recall e precisão para diferentes limites.

A implementação da curva Precision-Recall em sklearn (`precision_recall_curve`) não suporta diretamente problemas de classificação multiclasse por causa dessas complicações.

Para métricas multiclasse, contudo, precisamos de uma versão ampliada de Precisão e Recall. Existem duas abordagens principais: micro-average e macro-average.

Na micro-average, computamos a métrica globalmente contabilizando o total de verdadeiros positivos, falsos negativos e falsos positivos, ignorando a estrutura de classificação multiclasse e tratando cada classificação individualmente. Por outro lado, a macro-average calcula a métrica para cada classe individualmente e depois tira a média.

Então, para aplicar a curva Precision-Recall em problemas de classificação multiclasse, podemos calcular as métricas e plotá-las individualmente para cada classe. No entanto, essa abordagem não é diretamente suportada pela função `precision_recall_curve` do sklearn, o que requer implementação personalizada.

É por isso que o uso de `precision_recall_curve` neste cenário não faz sentido, dado que estamos lidando com um problema de classificação multiclasse. Uma alternativa mais adequada seria usar o relatório de classificação (`classification_report`) que fornece precisão, recall e f1-score para cada classe individualmente.

Se ainda desejarmos traçar uma curva de precisão e recall, podemos simplificar ainda mais o problema transformando o 'rating' em apenas duas classes: 'ruim' e 'bom'. Isso é um exemplo bastante característico de como as séries são avaliadas em plataformas como a Netflix, onde os usuários sinalizam se gostaram ou não do conteúdo. Para isso, iremos dividir novamente a coluna 'rating', usando 50 como critério: notas acima de 50 serão consideradas 'boas' e as demais, 'ruins'.



```python
data_case_two = data
# Transforma 'rating' em categorias
data_case_two['rating_category'] = (data_case_two['rating'] > 50).astype(int)
data_case_two
```

```python
# Divide os dados
train, test = train_test_split(data_case_two, test_size=0.2, random_state=123)

# Setup do ambiente de Classificação
exp_clf101 = setup(ignore_features= ['rating'], data = train, target = 'rating_category', session_id=123)

# Compara todos os modelos para encontrar o melhor
best_model = compare_models()

# Obtem as previsões nos dados de teste
predictions = predict_model(best_model, data = test)
```

```python
predictions
```

```python
# calcular a matriz de confusão
cm = confusion_matrix(test['rating_category'], predictions['prediction_label'])

# criar um DataFrame a partir da matriz de confusão
cm_df = pd.DataFrame(cm, index = ['ruim', 'bom'], columns = ['ruim', 'bom'])

# criar um heatmap a partir do DataFrame
plt.figure(figsize = (8, 6))
sns.heatmap(cm_df, annot = True, cmap='Blues', fmt = 'g')

# adicionar título e rótulos aos eixos
plt.title('Matriz de Confusão')
plt.xlabel('Valor Verdadeiro')
plt.ylabel('Valor Previsto')

# exibir o gráfico
plt.show()
```

```python
# Relatório de classificação
print(classification_report(test['rating_category'], predictions['prediction_label']))
```

```python
# Calcular Precisão e Recall
prec, recall, _ = precision_recall_curve(test['rating_category'], predictions.prediction_score)

# Calcular a AUC
auc_score = auc(recall, prec)

# Estilizar o plano de fundo do gráfico usando Seaborn
sns.set_style("whitegrid")

# Criar o gráfico usando Matplotlib
plt.figure(figsize=(10,6))
plt.plot(recall, prec, marker='.', label=f' AUC={auc_score:.2f}')
plt.title('Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend()

# Mostrar o gráfico
plt.show()
```
